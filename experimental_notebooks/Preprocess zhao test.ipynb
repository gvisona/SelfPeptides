```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = []
folder = "../data/zhao_binding_test"

for peptide_len in [9,10]:
    pfolder = os.path.join(folder, f"{peptide_len}mer")
    for fname in os.listdir(pfolder):
        if "full" in fname:
            continue
        # print(fname)
        allele = fname.replace(".txt", "").replace("%2A", "").replace("%3A", ":")
        allele_data = pd.read_csv(os.path.join(pfolder, fname), sep="\t", header=None)
        allele_data.columns = ["Peptide", "Affinity"]
        allele_data["Label"] = (allele_data["Affinity"]<500).astype(int)
        allele_data.insert(0, "HLA", allele)
        data.append(allele_data)
data = pd.concat(data)
data = data.drop("Affinity", axis=1)
data
```

```python
ps_df = pd.read_csv("/home/gvisona/Projects/SelfPeptides/processed_data/HLA_embeddings/HLA_pseudoseqs_T5/hla_pseudoseq_mapping.csv")
ps_df
```

```python
ps_mapping = {r[0]: r[1] for i,r in ps_df.iterrows()}
data["Allele Pseudo-sequence"] = data["HLA"].map(ps_mapping)
data
```

```python
prot_seq_file = pd.read_csv("/home/gvisona/Projects/SelfPeptides/processed_data/HLA_embeddings/HLA_proteins_T5/hla_proteins_mapping.csv")
prot_seq_file
```

```python
pseq_mapping = {r[1]: r[3] for i,r in prot_seq_file.iterrows()}
data["Allele Protein sequence"] = data["HLA"].map(pseq_mapping)
data
```

```python
data.to_csv("../processed_data/Binding_Affinity/Zhao_Benchmarking.csv", index=False)
```
