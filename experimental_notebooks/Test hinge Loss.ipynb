```python
import torch
import numpy as np

import torch.nn as nn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
xx = torch.tensor(np.linspace(-2, 2, 100))
```

```python
cmap = plt.get_cmap("inferno", 10)
```

```python
cmap(11)
```

```python
label = torch.ones(len(xx))

margins = np.linspace(0.0, 1.0, 10)
fig, ax = plt.subplots(figsize=(10, 7))

for i, m in enumerate(margins):
    loss = nn.HingeEmbeddingLoss(margin=m, reduction="none")
    yy = loss(xx, label)
    color = cmap(i)

    ax.plot(xx.numpy(), yy.numpy(), label="Margin {}".format(m))
ax.legend()
ax.set_title("Label {}".format(label[0].item()))
plt.show()
```

```python
label = torch.ones(len(xx))

margins = np.linspace(0.0, 1.0, 10)
fig, ax = plt.subplots(figsize=(10, 7))

label = +1
for i, m in enumerate(margins):
    # loss = max(0, m - label * xx)
    yy = torch.maximum(torch.tensor([0]), m - label * xx)
    color = cmap(i)

    ax.plot(xx.numpy(), yy.numpy(), label="Margin {}".format(m))
ax.legend()
ax.set_title("Label {}".format(label))
plt.show()
```

```python

margins = np.linspace(0.1, 1.0, 10)
fig, ax = plt.subplots(figsize=(10, 7))

label = -1
for i, m in enumerate(margins):
    # loss = max(0, m - label * xx)
    yy = torch.maximum(torch.tensor([0]), m - label * xx)
    color = cmap(i)

    ax.plot(xx.numpy(), yy.numpy(), label="Margin {}".format(m), c=color)
ax.legend()
ax.set_title("Label {}".format(label))
plt.show()
```
